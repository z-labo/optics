## 2. ファイバーの状態確認

LEDを位置1および位置2に設置し、カメラ1とカメラ2を用いて、それぞれのファイバー断面画像を記録する。

### STEP 1. LEDライトを位置1に置く

![Fig. 2](../.figure/MMFmanual/prep/11.png)

### STEP 2. カメラ2でファイバーの入力面を撮影する

無線キーボードとマウスをルームに持っていく。ルームに設置されているモニターから **StViewer_x64_v1_2** を起動し、**STC-MBS500U3V(26C4493)** カメラを選択する。その後、スタートボタンを押してライブ画像を確認する。

ファイバー断面を見つけやすくするため、右パネルで以下のように設定する。

**Remote Device → Acquisition Control → Exposure Mode → Timed → Exposure Time Raw → 2048**

![Fig. 1](../.figure/MMFmanual/prep/1.png)

![Fig. 2](../.figure/MMFmanual/prep/4.png)

ファイバー断面が最も明瞭に見えるように、露光時間と対物レンズの焦点位置を調整する。Fig. 3のような画像が確認できたら、画像名を **proximal** として保存する。

![Fig. 3](../.figure/MMFmanual/prep/5.png)

保存した画像をFig. 4に示すパスへコピーし、ファイル名を **reference** に変更する。

![Fig. 4](../.figure/MMFmanual/prep/6.png)

### STEP 3. LEDライトを位置2に置く

![Fig. 4](../.figure/MMFmanual/prep/8.png)

### STEP 4. カメラ1でファイバーの出力面を撮影する

今回はファイバーの出力断面を撮影する。同じ手順で、**STC-MBS500U3V(25E2986)** カメラを選択し、最も焦点が合う面で撮影する。画像名を **distal** として保存する。

## 3. SLMのキャリブレーションを行う

SLMとファイバー入力面の相対位置は、温度変化や微小な空気の流れによって変化する。そのため、キャリブレーションに用いる集光スポットの位置を正確に制御するには、SLMとファイバー入力断面の位置関係をキャリブレーションする必要がある。

### STEP 1. Calibration GUIを開く

VS Codeを起動し、File → Open Folder から、Fig. 5のように **MMF_Experiment** フォルダーを選択する。

![Fig. 5](../.figure/MMFmanual/calibration/2.png)

左パネルから `calibration_gui.py` を選択する。その後、Fig. 6のように、VS Code画面右下にある **Interpreter** と **Virtual Environment** をそれぞれ設定する。

![Fig. 6](../.figure/MMFmanual/calibration/3.png)

InterpreterはFig. 7のように選択する。

![Fig. 7](../.figure/MMFmanual/calibration/4.png)

Virtual EnvironmentはFig. 8のように選択する。

![Fig. 8](../.figure/MMFmanual/calibration/6.png)

### STEP 2. 3つの集光スポットを用いて位置をキャリブレーションする

環境選択が完了すると、Fig. 9のようにコマンドプロンプトに `(.venv)` が表示される。

![Fig. 9](../.figure/MMFmanual/calibration/8.png)

`calibration_gui.py` を実行する。右上の実行ボタンを使用してもよいし、コマンドプロンプトから以下のコマンドを入力して実行してもよい。

```bash
python ./calibration.py
```

実行すると、Fig. 10のようなGraphical User Interfaceが表示される。

![Fig. 10](../.figure/MMFmanual/calibration/10.png)

右側の **Calibration by picking 3 points** パネルから **Pick 3 spots** を押すと、自動的にカーソルが入力待ち状態になる。左側の **Proximal** 画像上で、なるべく縁に近く、互いに離れた3点をマウスでクリックする。クリックした3点の座標はPython側で記録される。

同時に、最後に保存されたキャリブレーションファイルが読み込まれ、SLMパターンが作成・表示される。そのため、Fig. 11のように、カメラ2を用いてファイバー入力面に入射する光の反射成分をモニターできる。

![Fig. 11](../.figure/MMFmanual/calibration/13.png)

集光スポットがぎりぎり確認できる程度に露光時間を調整し、手動で画像を保存する。

次に、Fijiを開き、保存したキャリブレーション画像をドラッグアンドドロップで読み込む。ピクセル選択ツールを用いて、3つの集光スポットの座標を読み取る。具体的な手順は以下の図を参照する。

![Fig. 12](../.figure/MMFmanual/calibration/15.png)

![Fig. 13](../.figure/MMFmanual/calibration/16.png)

![Fig. 14](../.figure/MMFmanual/calibration/17.png)

Fijiで読み取った実際の座標をGUIに入力し、**Save calibration** ボタンを押す。保存が完了したら、一度GUIを閉じる。

## 4. MMFキャリブレーションを行う

### STEP 1. スキャン領域を決める

GUIを再度起動する。先ほど保存したキャリブレーションデータは、自動的に反映された状態になっている。

Fijiで **proximal** 画像データを開き、正方形選択ツールを使ってファイバー断面を正方形で囲む。このとき、左上のX, Y座標を読み取り、GUIの `lx1`, `ly1` に入力する。また、右下のX, Y座標を読み取り、`lx2`, `ly2` に入力する。

![Fig. 15](../.figure/MMFmanual/calibration/19.png)

![Fig. 16](../.figure/MMFmanual/calibration/20.png)

### STEP 2. スキャンパターンを作成し、キャリブレーションデータを取得する

その後、**Scan grid** ボタンを押すと、SLMパターンが計算され、SLMリストに格納される。

**Camera capture off** の状態でカメラ1とカメラ2を起動し、それぞれの露光時間を以下のように設定する。

- カメラ1：512
- カメラ2：10

また、SLMリスト内のパターンを選択すると、事前に集光スポットを確認できる。画像内にサチュレーションしている領域がないか確認する。

![Fig. 17](../.figure/MMFmanual/calibration/22.png)

露光時間の設定が完了したら、StViewerを両方とも閉じる。その後、ルームのモニターの電源を切り、キーボードとマウスを外に持ち出す。ルームのドアを閉める際には、実験系に衝撃を与えないよう注意する。

SLMリストの下にある **Camera capture** のチェックボックスにチェックを入れ、**Play SLM** を押すと記録が開始される。すべてのSLMパターンに対して、カメラ1とカメラ2でそれぞれ1枚ずつ画像が取得される。このとき、露光時間などの設定は、最後にStViewerで設定した値が使用される。

キャリブレーションデータは、現在の日付と時刻をフォルダー名として、以下のパスに自動保存される。準備段階で取得した **distal** および **proximal** の画像も、このフォルダー内に保存する。

64×64パターンの場合、撮影完了まで約15分かかる。**display and capture OK** というメッセージが表示されたら、GUIを閉じてもよい。

![Fig. 18](../.figure/MMFmanual/calibration/24.png)

## 5. 透過行列を計算し、出力側で任意のパターンを作成する

### STEP 1. 画像の前処理を行う

`pipeline_GUI.py` ファイルを実行し、**Save Base Directory** に、キャリブレーション時に自動生成されたフォルダーのパスをコピー＆ペーストする。

![Fig. 18](../.figure/MMFmanual/pipeline/2.png)

その後、**Check off-aixs hologram** ボタンを押すと、しばらくしてポップアップ画面が表示される。最初に表示されるポップアップは、出力画像を処理し、断面を円形にクロップした結果を示している。内容を確認してから閉じる。

続いて2番目のポップアップが表示される。この画面は、オフアクシス型デジタルホログラフィの処理条件を決めるためのステップであるため、慎重に確認する必要がある。特に、一番左に表示される空間周波数画像には3つの信号成分が現れる。その中で、一番下、または右下にある信号成分を抽出して位相を計算する必要がある。

抽出領域が正しく設定されているかを確認し、マスク位置が少しでもずれている場合は、画像を拡大して信号成分の中心座標を正確に読み取り、GUIに入力する。ただし、入力する際には、読み取った座標の順番を逆にする。

![Fig. 19](../.figure/MMFmanual/pipeline/5.png)

![Fig. 20](../.figure/MMFmanual/pipeline/6.png)

![Fig. 21](../.figure/MMFmanual/pipeline/7.png)

中心座標を入力したら、もう一度 **Check off-aixs hologram** ボタンを押して結果を確認する。このとき、中心座標はファイルとして保存され、次回以降は自動的に読み込まれる。

次に、**Read experiment Data** を押して、すべての実験データに対して自動で前処理を行う。

### STEP 2. 透過行列を求め、出力側で走査する集光スポットを形成する

**Calculate input fields** を押すと、出力側で集光スポットを形成するためのSLMパターンが自動的に計算され、ファイルとして保存される。

先にカメラ1のViewerを起動してから、**Display on SLM** を押す。保存されたSLMパターンが順番に表示され、出力側で集光点が走査される様子をライブ画像で確認できる。

**On Camera** にチェックを入れてから **Display on SLM** を押すと、ライブ表示ではなく、画像が自動的に保存される。

### STEP 3. 手書きで任意のパターンを設計し、そのパターンを表示する

**Run demonstration** を押すと、手書き可能な画面が表示される。任意のパターンを描画して **Save** を押し、画面を閉じると、約1分で表示用のSLMパターンが計算され、表示される。ライブカメラでその結果を確認する。

![Fig. 23](../.figure/MMFmanual/pipeline/14.png)
