## 2. ファイバーの状態確認

LEDを位置１と位置２において、カメラ１とカメラ２でそれぞれのファイバー断面の画像を記録する。

### STEP 1. LEDライトを位置１に置く
![Fig. 2](../.figure/MMFmanual/prep/11.png)

### STEP 2. カメラ２でファイバーの入力面を撮影
無線キーボードとマウスをAKルームにもっていく。AKルームに用意されているモニターから**StViewer_x64_v1_2**を起動、**STC-MBS500U3V(26C4493)**カメラを選択して、スタートボタンを押して、ライブ画像を確認する。その時、右パネルのRemote Device → Acquisition Control → Exposure Mode → Timed → Exposure Time Raw → 2048にすると見つけやすい。

![Fig. 1](../.figure/MMFmanual/prep/1.png)

![Fig. 2](../.figure/MMFmanual/prep/4.png)

ファイバーの断面がもっともはっきり見えるように、露光時間を調整、対物レンズの焦点距離を調整すると、Fig. 3のような画像が確認できる。この画像を名前「proximal」と保存ずる。

![Fig. 3](../.figure/MMFmanual/prep/5.png)!

保存した画像をFig. 4のPathにコピーして、ファイル名を**reference**に変える。

![Fig. 4](../.figure/MMFmanual/prep/6.png)!

### STEP 3. LEDライトを位置2に置く

![Fig. 4](../.figure/MMFmanual/prep/8.png)!

### STEP 4. カメラ1でファイバーの入力面を撮影
今回はファイバーの出力断面を撮影する。同じ手順で、**STC-MBS500U3V(25E2986)**カメラから一番焦点が合う面で撮影して、名前「distal」と保存ずる。

## 3. SLMのキャリブレーションを行う
SLMとファイバーの入力面の相対位置は、温度変化や、微小な空気の流動によって変化する。キャリブレーションに使う集光スポットの位置を正確に制御するために、SLMと入力断面位置のキャリブレーションを行う。

### calibration GUIを開く
VS codeを立ち上げ、File → Open Flolder からFig. 5のように**MMF_Experiment**フォルダーを選択する。

![Fig. 5](../.figure/MMFmanual/calibration/2.png)!

左パネルからcalibration_gui.pyを選択した後に、Fig. 6のように、VS Code画面の右下にある、EnterpreterとVirtual Envaronmentをそれぞれ設定する。

![Fig. 6](../.figure/MMFmanual/calibration/3.png)!

EnterpreterはFig. 7のように選択する。

![Fig. 7](../.figure/MMFmanual/calibration/4.png)!

Virtual EnvaronmentはFig. 8のように選択する。

![Fig. 8](../.figure/MMFmanual/calibration/6.png)!

### 集光スポット３つから位置をキャリブレーションする

環境選択が完了したらFig. 9のようにコマンドプロンプトに(.venv)が付くことが確認できる。
![Fig. 9](../.figure/MMFmanual/calibration/8.png)!
calibration_gui.pyを実行する。右上の実行ボタンを使ってもいいし、コマンドプロンプトから　python ./calibration.pyを入力して実行してもよい。
Fig. 10のようなGuide User interfaceがポップアップする。

![Fig. 10](../.figure/MMFmanual/calibration/10.png)!

右の**calibration by picking 3 points**パネルから**Pick 3 spots**を押すと、自動でカーソルが入力待ち状態になり、左のProsimalの画像からなるべく縁の異なる３点をマウスクリックすると、その座標をPythonが記憶する。同時に、最後に行ったキャリブレーションのファイルを読み込みSLMパターンを作成し、表示するので、Fig. 11のようにカメラ２でファイバーの入力面に入る光の反射光成分をモニターすることができる。

![Fig. 11](../.figure/MMFmanual/calibration/13.png)!

この画像を集光スポットがギリギリ見えるように露光時間を調整して、手動で画像を保存する。
次にFijiソフトウエアを開いて、保存したキャリブレーション画像をドラッグアンドドロップで開き、ピクセル選択ツールを使って３点の座標を読み取る。具体的はやり方は下のFigsを参照するとよい

![Fig. 12](../.figure/MMFmanual/calibration/15.png)!

![Fig. 13](../.figure/MMFmanual/calibration/16.png)!

![Fig. 14](../.figure/MMFmanual/calibration/17.png)!

読み取った実際の座標を、GUIに入力して、Save calibrationボタンを押す。これが終わったら一度GUIを閉じる。

##　4. ファイバー入力面をスキャンして、透過行列を求めるためのデータをとる。(MMFのキャリブレーション)

### スキャン領域を決める

GUIを再度立ち上げる。キャリブレーションデータは先ほど保存したデータがすでに反映された状態である。Fijiからdistal画像データを読み、正方形選択ボタンを使って、断面を正方形に囲む。その時、左上のX,Y座標を読んで、lx1, ly1に入力し、右下の座標を読んでlx2,ly2に入力してから**scan grid**ボタンを押すと、SLMパターンが計算されて、SLMリストに格納される。

![Fig. 15](../.figure/MMFmanual/calibration/19.png)!

![Fig. 16](../.figure/MMFmanual/calibration/20.png)!

Camera capure offの状態でカメラ１と２を立ち上げ、それぞれの露光時間を512と10に設定する。また、SLMリストのパターンを選択すると、事前に集光スポットを確認することができる。画像にサチュレーションしたどころがないか確認する。

![Fig. 17](../.figure/MMFmanual/calibration/22.png)!

露光時間の設定が終わったら、StViewerを両方閉じて、AKルームのモニターの電源を切り、キーボードとマウスを持ち出して、外で作業をする。AKルームのドアを閉じるときには、衝撃を与えないようにする。
SLMリストの下にあるCamera capureのチェックボックスにチェックを入れてから、Play SLMを押すと、記録が始まる。すべてのSLMパターンに対して、カメラ1とカメラ２で1枚ずつ撮影する。その時露光時間などの設定は最後にStviewerで行った設定を使用する。
キャリブレーションデータは自動で現在の日付と時間をフォルダー名として、以下のPathに保存される。準備段階でとった、DistalとProximalの画像もこのフォルダの中に収める。64x64パターンで撮影が完了するまで約15分程度かかる。display and capture OKメッセージが出るとGUIを閉じてもよい。

![Fig. 18](../.figure/MMFmanual/calibration/24.png)!
