## 3. スポット位置のキャリブレーションを行う

SLMとファイバー入力面の相対位置は、温度変化や微小な空気の流れによって変化する。

そのため、キャリブレーションに用いる集光スポットの位置を正確に制御するには、SLMとファイバー入力断面の位置関係をキャリブレーションする必要がある。

VS Codeを起動し、File → Open Folder から、Fig.8のように **MMF_Experiment** フォルダーを選択する。

| ![](../.figure/MMFmanual/calibration/1.png) |![](../.figure/MMFmanual/calibration/2.png) |
|---|---|
| Fig.8 Open folder|Fig.9 Find MMF_Experiment folder from my document|

フォルダーが選択されたら、左パネルから `calibration_gui.py` を選択する。Fig.9のようにコマンドプロンプトに `(.venv)` が表示されるのを確認してから、コマンドプロンプトに**python ./calibration.py**を入力してプログラムを実行する。
| ![](../.figure/MMFmanual/calibration/8.png) |
|---|
| Fig.9 Type in command line to run the program|

```bash
python ./calibration.py
```

Graphical User Interface(GUI)が表示されると、右側の **Calibration by picking 3 points** パネルから **Pick 3 spots** を押す。
| ![](../.figure/MMFmanual/calibration/11.png) |
|---|
| Fig.10 Pick 3 points to calibrate spot position |

カーソルが自動的に入力待ち状態に変わる。

左側に読み込まれた**reference.bmp**画像上で、なるべく縁に近く、互いに離れた3点をマウスでクリックする。クリックした3点の座標はPython側で記録される。

同時に、最後に保存されたキャリブレーションデータが読み込まれ、SLMパターンが作成・表示される。

その結果、Fig.11のように、カメラ1を用いてファイバー入力面に入射する光の反射成分をモニターできる。
| ![](../.figure/MMFmanual/calibration/13.png) |
|---|
| Fig.11 Monitor input spots|

集光スポットがぎりぎり確認できる程度に**露光時間**を調整し、手動で画像を保存する。

次に、Fiji(Fig.12)を開き、保存したキャリブレーション画像をドラッグアンドドロップで読み込む。ピクセル選択ツールを用いて、3つの集光スポットの座標を読み取る(Fig.13)。

| ![](../.figure/MMFmanual/calibration/15.png) |![](../.figure/MMFmanual/calibration/17.png) |
|---|---|
| Fig.12 Open Fiji|Fig.13 read spot positions in pexels|

Fijiで読み取った実際の座標をGUIに入力し、**Save calibration** ボタンを押す。

保存が完了したら、一度GUIを閉じる。


## 4. MMF透過行列を計算するための、キャリブレーションデータを取得する
### 4.1. スキャン領域を決める

GUIを再度起動する。

先ほど保存したキャリブレーションデータは、自動的に反映された状態になっている。

Fijiで **reference.bmp** 画像データを開き、正方形選択ツールを使ってファイバー断面を正方形で囲む。
このとき、左上のX, Y座標を読み取り、GUIの `lx1`, `ly1` に入力する。
また、右下のX, Y座標を読み取り、`lx2`, `ly2` に入力する。

| ![](../.figure/MMFmanual/calibration/19.png) |![](../.figure/MMFmanual/calibration/20.png) |
|---|---|
| Fig.14 read scan area|Fig.15 type in the area|


### 4.2. スキャンパターンを作成し、キャリブレーションデータを取得する

その後、**Scan grid** ボタンを押すと、SLMパターンが計算され、SLMリストに格納される。

**Camera capture** にチェックが入ってない状態でカメラ1とカメラ2を起動し、それぞれの露光時間を以下のように設定する。

- カメラ1：10
- カメラ2：512

| ![](../.figure/MMFmanual/calibration/22.png) |
|---|
| Fig.15 Open both cameras to see the images|

また、SLMリスト内の任意のパターンを選択すると、事前に集光スポットの状況を確認できる。

画像内にサチュレーションしている領域がないか確認する。

露光時間の設定が完了したら、StViewerを両方とも閉じる。その後、ダークルームのモニターの電源が切れていることを確認し、ドアを閉める。

SLMリストの下にある **Camera capture** のチェックボックスにチェックを入れ、**Play SLM** を押すと記録が開始される。

すべてのSLMパターンに対して、カメラ1とカメラ2でそれぞれ1枚ずつ画像が取得される。このとき、露光時間などの設定は、最後にStViewerで設定した値が使用される。

キャリブレーションデータは、現在の日付と時刻をフォルダー名として、以下のパスに自動保存される。準備段階で取得したカメラ１とカメラ２の断面の画像も、このフォルダー内に保存する。

64×64パターンの場合、撮影完了まで約15分かかる。**display and capture OK** というメッセージが表示されたら、GUIを閉じてもよい。


## 5. 透過行列を計算し、出力側で任意のパターンを作成する

### 5.1. 画像の前処理を行う

`pipeline_GUI.py` ファイルを実行し、**Save Base Directory** に、キャリブレーション時に自動生成されたフォルダーのパスをコピー＆ペーストする。

| ![](../.figure/MMFmanual/pipeline/2.png) |
|---|
| Fig.16 copy the experiment data folder to save base directory|

その後、**Check off-aixs hologram** ボタンを押すと、しばらくしてポップアップ画面が表示される。

最初に表示されるポップアップは、出力画像を処理し、断面を円形にクロップした結果を示している。内容を確認してから閉じる。

続いてFIg.17のように、2番目のポップアップが表示される。

この画面は、オフアクシス型デジタルホログラフィの処理条件を決めるためのステップであるため、**慎重に確認する必要がある**。

特に、一番左に表示される空間周波数画像には3つの信号成分の中で、下の方にある信号成分を抽出して位相を計算する必要があるので、その成分の中心点を読み、GUIに入力する。**ただし、入力する際には、読み取った座標の順番を逆にする**。

フィルターリングマスク位置が少しでもずれている場合は、Fig.18のように画像を拡大するなりして、信号成分の中心座標を正確に読み取りなおす。

| ![](../.figure/MMFmanual/pipeline/5.png) |![](../.figure/MMFmanual/pipeline/6.png) |
|---|---|
| Fig.17 extract phase signal|Fig.18 zoom in the image and read the pixels|


pipeline_guiに中心座標を入力したら、もう一度 **Check off-aixs hologram** ボタンを押して結果を確認する。

このとき、中心座標はファイルとして保存され、次回以降は自動的に読み込まれる。

次に、**Read experiment Data** を押して、すべての実験データに対して自動で前処理を行う。

### 5.2. 透過行列を求め、出力側で走査する集光スポットを形成する

**Calculate input fields** を押すと、出力側で集光スポットを形成するためのSLMパターンが自動的に計算され、ファイルとして保存される。

先にカメラ1のViewerを起動してから、**Display on SLM** を押す。保存されたSLMパターンが順番に表示され、出力側で集光点が走査される様子をライブ画像で確認できる。

**On Camera** にチェックを入れてから **Display on SLM** を押すと、ライブ表示ではなく、画像が自動的に保存される。

### 5.3. 手書きで任意のパターンを設計し、そのパターンを表示する

**Run demonstration** を押すと、手書き可能な画面が表示される。任意のパターンを描画して **Save** を押し、画面を閉じると、約1分で表示用のSLMパターンが計算され、表示される。ライブカメラでその結果を確認する。
