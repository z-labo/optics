# Multimode fiber experiment manual

## 1. 事前準備

実験の30分前にレーザーとSLMの電源をONにする

---

## 2. ファイバーの状態確認

LEDを位置１と位置２において、カメラ１とカメラ２でそれぞれのファイバー断面の画像を記録する。

### STEP 1. LEDライトを位置１に置く

### STEP 2. カメラ２でファイバーの入力面を撮影
無線キーボードとマウスをAKルームにもっていく。AKルームに用意されているモニターから**StViewer_x64_v1_2**を起動、**STC-MBS500U3V(26C4493)**カメラを選択して、スタートボタンを押して、ライブ画像を確認する。その時、右パネルのRemote Device → Acquisition Control → Exposure Mode → Timed → Exposure Time Raw → 2048にすると見つけやすい。

![Fig. 1](../.figure/MMFmanual/prep/1.png)

![Fig. 2](../.figure/MMFmanual/prep/4.png)

ファイバーの断面がもっともはっきり見えるように、露光時間を調整、対物レンズの焦点距離を調整すると、Fig. 3のような画像が確認できる。この画像を名前「proximal」と保存ずる。

![Fig. 3](../.figure/MMFmanual/prep/5.png)!

保存した画像をFig. 4のPathにコピーして、ファイル名を**reference**に変える。

![Fig. 4](../.figure/MMFmanual/prep/6.png)!

### STEP 3. LEDライトを位置2に置く

### STEP 4. カメラ1でファイバーの入力面を撮影
今回はファイバーの出力断面を撮影する。同じ手順で、**STC-MBS500U3V(25E2986)**カメラから一番焦点が合う面で撮影して、名前「distal」と保存ずる。

## 3. SLMのキャリブレーションを行う
SLMとファイバーの入力面の相対位置は、温度変化や、微小な空気の流動によって変化する。キャリブレーションに使う集光スポットの位置を正確に制御するために、SLMと入力断面位置のキャリブレーションを行う。

### calibration GUIを開く
VS codeを立ち上げ、File → Open Flolder からFig. 5のように**MMF_Experiment**フォルダーを選択する。

![Fig. 5](../.figure/MMFmanual/calibration/2.png)!

左パネルからcalibration_gui.pyを選択した後に、Fig. 6のように、VS Code画面の右下にある、EnterpreterとVirtual Envaronmentをそれぞれ設定する。

![Fig. 6](../.figure/MMFmanual/calibration/3.png)!

EnterpreterはFig. 7のように選択する。

![Fig. 7](../.figure/MMFmanual/calibration/4.png)!

Virtual EnvaronmentはFig. 8のように選択する。

![Fig. 8](../.figure/MMFmanual/calibration/6.png)!
