# TE mode and Bessel functions

## 1. 座標系と波数ベクトル

円筒座標を

$$
(r,\phi,z)
$$

とする。ここで

- $z$：導波管の軸方向
- $r$：軸からの距離
- $\phi$：横断面内の方位角

である。

導波管内を $z$ 方向へ伝搬する波では、波数ベクトルは軸方向成分と横方向成分に分けて

$$
\mathbf{k}=\mathbf{k}_\perp+\beta \hat z
$$

と書ける。円筒対称な議論では、横波数の大きさを $k_r$ と書き、

$$
k_r=nk_0\sin\theta,\qquad
\beta=nk_0\cos\theta
$$

とおくことができる。ここで $\theta$ は波数ベクトル $\mathbf{k}$ と $z$ 軸とのなす角である。したがって $\theta$ は、「波が導波管軸に対してどれだけ傾いているか」を表す。  

---

## 2. TE モードの意味

TE モードとは、軸方向電場成分がゼロであるモード、すなわち

$$
E_z=0
$$

を満たすモードである。ただし、これは「$z$ 方向に伝搬しない」という意味ではない。実際には場は

$$
e^{-i\beta z}
$$

のような $z$ 依存性を持ち、$\beta\neq 0$ ならば波は軸方向へ伝搬する。

したがって TE とは、「電場が完全に横方向だけを向いている」という意味であり、「伝搬方向が横方向だけである」という意味ではない。円形導波管では TE モードにおいて縦成分として残るのは一般に $H_z$ であり、この $H_z$ から横電場・横磁場が決まる。

---

## 3. 「反射する平面波の重ね合わせ」としての直感

導波管内のモードは、軸方向成分 $\beta$ を共有しつつ、横方向に異なる向きをもつ平面波成分の重ね合わせとして直感的に理解できる。1つの成分だけを見れば、たとえば

$$
\mathbf{k}=k_r\hat r+\beta \hat z
$$

のように書ける。均一媒質中の平面波では

$$
\mathbf{E}\cdot \mathbf{k}=0
$$

であるため、電場は $\mathbf{k}$ に垂直でなければならない。円筒座標の直感では、そのような横方向成分として $\hat\phi$ 方向の電場を考えると、$\hat r$ にも $\hat z$ にも直交するので、TE 的な偏光をイメージしやすい。

さらに壁で反射すると、横波数成分の符号は反転して

$$
k_r\rightarrow -k_r
$$

となるが、$\beta$ は変わらない。したがって、1次元的な直感では

$$
e^{i(k_r r+\beta z)}+e^{i(-k_r r+\beta z)}
=
2\cos(k_r r)e^{i\beta z}
$$

のように、$\pm k_r$ 成分の重ね合わせから横方向の定在波構造が生じる。この意味で

$$
\langle k_r\rangle=0,\qquad k_r\neq 0
$$

である。すなわち、**平均として横方向へ流れてはいないが、横方向スケールそのものは残っており、その結果として横断面モード構造が形成される**。

ただし、これはあくまで 1 次元的直感であり、円筒導波管の厳密な半径方向関数は $\cos(k_r r)$ ではなく、最終的にはベッセル関数 $J_l(k_r r)$ になる。  

---

## 4. 円筒対称性とベッセル関数の出現

円筒系では、ある方位角 $\psi$ の方向を向く横波数ベクトルをもつ平面波成分は、観測点 $(r,\phi)$ において

$$
e^{i k_r r\cos(\phi-\psi)}
$$

という位相因子を与える。これを全方位について等重みで重ね合わせると

$$
\int_0^{2\pi} e^{i k_r r\cos(\phi-\psi)}\,d\psi
$$

となる。

ベッセル関数の積分表示

$$
J_0(k_r r)
=
\frac{1}{2\pi}
\int_0^{2\pi}
e^{i k_r r\cos(\phi-\psi)}\,d\psi
$$

より、

$$
\boxed{
\int_0^{2\pi}
e^{i k_r r\cos(\phi-\psi)}\,d\psi
=
2\pi J_0(k_r r)
}
$$

が得られる。したがって、**円筒対称な平面波重ね合わせから $J_0$ が自然に現れる**。
さらに、方位角モード数 $l$ をもつ成分を作るには、各平面波成分に対して方位角 $\psi$ に依存する位相重み

$$
e^{il\psi}
$$

を掛けて重ね合わせればよい。その結果、

$$
\boxed{
\int_0^{2\pi}
e^{il\psi}
e^{i k_r r\cos(\phi-\psi)}
\,d\psi
=
2\pi i^l J_l(k_r r)e^{il\phi}
}
$$

となる。これは Jacobi–Anger 展開

$$
e^{iz\cos\theta}
=
\sum_{m=-\infty}^{\infty}
i^m J_m(z)e^{im\theta}
$$

の直接の帰結であり、**平面波の重ね合わせが円筒波 $J_l(k_r r)e^{il\phi}$ を生む**ことを示している。

## 5. 円形導波管の TE モードの一般形

円形導波管の TE モードでは $E_z=0$ であり、縦磁場 $H_z$ が基本となる。一般に TE モードの縦磁場は

$$
H_z(r,\phi,z)
=
H_0\,J_l(k_c r)\,[A\cos(l\phi)+B\sin(l\phi)]\,e^{-i\beta z}
$$

の形で書ける。複素表示を用いれば

$$
H_z(r,\phi,z)\propto J_l(k_c r)e^{il\phi}e^{-i\beta z}
$$

と書いてもよい。ここで $k_c$ は横方向の固有波数であり、境界条件によって量子化される。

TE モードでは導体壁 $r=a$ において接線電場がゼロでなければならないため、境界条件は

$$
J_l'(k_c a)=0
$$

となる。したがって TE モードの固有値は、ベッセル関数そのものの零点ではなく、**その導関数の零点**によって決まる。
なお、円形導波管の支配モード（最も低い遮断周波数をもつモード）は $TE_{11}$ である。

---

## 6. $TE_{0m}$ において $E_\phi\propto J_1$ が現れる理由

軸対称な TE モードでは $l=0$ であるから、縦磁場は

$$
H_z(r,z)\propto J_0(k_c r)e^{-i\beta z}
$$

となる。

TE モードの横電場成分は Maxwell 方程式から $H_z$ の半径方向微分で決まり、特に軸対称の場合には

$$
E_\phi \propto -\frac{\partial H_z}{\partial r}
$$

となる。したがって

$$
E_\phi
\propto
-\frac{\partial}{\partial r}J_0(k_c r)
$$

である。ここでベッセル関数の微分公式

$$
\frac{dJ_0(x)}{dx}=-J_1(x)
$$

を使うと

$$
E_\phi\propto J_1(k_c r)
$$
を得る。ゆえに、$TE_{0m}$ では縦磁場が $J_0$ 型であっても、横電場 $E_\phi$ は $J_1$ 型になる。

---

## 7. $l$ の意味

モードの複素表示

$$
E(r,\phi,z)\propto J_l(k_r r)e^{il\phi}e^{-i\beta z}
$$

において、$l$ は **方位角モード数**である。

- $l=0$ なら $e^{i0\phi}=1$ であり、場は方位角 $\phi$ に依存しない。
- $l=1$ なら $e^{i\phi}$ であり、$\phi$ を $2\pi$ だけ回すと複素位相は $2\pi$ 変化する。
- $l=2$ なら $e^{i2\phi}$ であり、1周で複素位相は $4\pi$ 変化する。

この意味で、$l$ は複素表示では「位相の巻き数」に対応して見える。しかし、円形導波管の通常の TE/TM モードでは、しばしば

$$
\cos(l\phi),\qquad \sin(l\phi)
$$

という実関数表示で書かれるので、最も安全な呼び方は **方位角モード数**である。すなわち、$l$ は**偏光方向**を表す量ではなく、**横断面内での角度依存の次数**を表す量である。  

---

## 8. まとめ

以上をまとめると、円形導波管の TE モードは

$$
E_z=0,\qquad H_z\neq 0
$$

を満たすモードであり、横方向構造は平面波成分の重ね合わせとして理解できる。その円筒対称な重ね合わせの結果として、半径方向依存はベッセル関数 $J_l$ で表される。特に、等方位重ね合わせからは $J_0$ が現れ、方位角に位相重み $e^{il\psi}$ を与えると $J_l(k_r r)e^{il\phi}$ が現れる。円形導波管の TE モードでは縦磁場 $H_z$ が基本となり、$TE_{0m}$ では $H_z\propto J_0$、そこから導かれる横電場は

$$
E_\phi\propto J_1
$$

となる。したがって、$l$ は偏光を表すのではなく、**モードの方位角依存性を表す指標**である。
