# ファイバーからの出射角

## 平坦な端面におけるNAの計算

平坦な出射端面をもつファイバーコア内を伝搬する光を考える。
\
![Fig. 1](../.figure/fig1.jpg)

\
全反射条件を満たすためには、内部入射角 $\theta_1$ は臨界角 $\theta_c$ より大きくなければならない。このとき、出射境界面での屈折角 $\phi_1$ は幾何学的に次のように与えられる。

$$
\phi_1 = \frac{\pi}{2} - \theta_1
$$

境界面におけるスネルの法則 $n_1 \sin \phi_1 = n_0 \sin \phi_0$ に上式を代入すると、

$$
n_1 \cos \theta_1 = n_0 \sin \phi_0
$$

となる。したがって、$\phi_0$ の最大値は $\theta_1 = \theta_c$ のときに得られ、

$$
\sin \phi_0^{\max} = \frac{n_1}{n_0} \cos \theta_c
$$

臨界角の定義 $\sin \theta_c = \frac{n_2}{n_1}$ を用いると、

$$
\cos \theta_c = \sqrt{1 - \left(\frac{n_2}{n_2}\right)^2}
$$

最終的に、

$$
\sin \phi_0^{\max} = \frac{\sqrt{n_1^2 - n_0^2}}{n_0}
$$

となる。

## 角度をもつ端面 — ケース1

次に問題を少し変形し、出射面が平坦ではなく、角度 $\beta$ をもつテーパー端面であると仮定する。
![Fig. 2](../.figure/fig2.jpg)

まず、コアの中心に沿って直進する光を考えると、境界面では

$$
n_1 \sin\phi_1 = n_0 \sin\phi_0
$$

が成り立つ。$\phi_1 = \dfrac{\pi}{2} - \beta$、$\theta_0 = \phi_0 - \phi_1$ を代入すると、

$$
\theta_0 = \sin^{-1}\!\left(\frac{n_1}{n_0}\cos\beta\right) - \left(\frac{\pi}{2} - \beta\right)
$$

が導かれる。ただし、出射するためには、コア–外部媒質の境界面において $\phi_1$ が臨界角 $\theta_c$ より小さくなければならない。したがって、$\beta$ は $\dfrac{\pi}{2} - \theta_c$ より大きくなければならない。

## 角度をもつ端面 — ケース2

Fig. 2をさらに変形し、全反射しながら出射する光を考える。
![Fig. 3](../.figure/fig3.jpg)

同様に、出射境界面では $n_1 \sin\phi_1 = n_0 \sin\phi_0$ が成り立つ。ただし今回は、$\phi_1$ の値は

$$
\phi_1 = \frac{\pi}{2} - \beta - (\frac {\pi}{2} - \theta_1) = \theta_1 - \beta
$$

となるため、

$$
\theta_0 = \phi_0 - \phi_1
$$

に代入すると、

$$
\theta_0 = sin^{-1}[\frac{n_1}{n_0} sin(\theta_1 - \beta)] - (\theta_1 - \beta)
$$

が導かれる。ただし、ここでは2つの境界面における全反射条件を満たす必要がある。
1. $\theta_1$ はコア–クラッド境界面における臨界角 $\theta_{c1}$ より大きくなければならない。すなわち、$\theta_1 \ge sin^{-1}(\frac{n_2}{n_1})$。
2. $\phi_1$ はコア–外部媒質境界面における臨界角 $\theta_{c2}$ より小さくなければならない。すなわち、$\theta_1-\beta<sin^{-1}(\frac{n_0}{n_1})$。


## 角度をもつ端面 — ケース3

Fig. 3をさらに変形し、全反射しながら出射する光を考える。

![Fig. 4](../.figure/fig9.jpg)

出射境界面では $n_2 \sin\theta_2 = n_0 \sin\theta_0$ が成り立つ。ここでは、コアとクラッドの間の屈折を一時的に無視する。

今回は、$\theta_2$ の値は

$$
\theta_2 = \pi - \beta - (\frac{\pi}{2} - \phi_1) = \frac{\pi}{2} - \beta  + \phi_1 
$$

となる。これを下式に代入すると、

$$
\theta_0 = sin^{-1}[\frac{n_2}{n_0} cos(\beta - \phi_1)]
$$

が得られる。また、$\phi_1 = \theta_1 - \beta$ を代入すると、

$$
\theta_0 = sin^{-1}[\frac{n_2}{n_0} cos(2\beta - \theta_1)]
$$

が導かれる。ただし、ここでは2つの境界面における全反射条件を満たす必要がある。
1. $\theta_1$ はコア–クラッド境界面における臨界角 $\theta_{c1}$ より大きくなければならない。すなわち、$\theta_1 \ge sin^{-1}(\frac{n_2}{n_1})$。
2. $\phi_1$ はコア–外部媒質境界面における臨界角 $\theta_{c2}$ より大きくなければならない。すなわち、$\phi_1 \ge sin^{-1}(\frac{n_1}{n_0})$。

```python
import numpy as np
import matplotlib.pyplot as plt

# パラメータ
n1 = 1.447
n0 = 1.0   # 空気 

# 臨界角
theta_c = np.arcsin(n0 / n1)

# カットオフとなる beta（出射開始点）
beta_cutoff = np.pi/2 - theta_c

# beta の範囲（0～90度）
beta = np.linspace(0, np.pi/2, 1000)

# 定義域を確認
argument = (n1 / n0) * np.cos(beta)

# 有効領域のマスク
valid = argument <= 1

# theta0 を計算（無効値は NaN）
theta0 = np.full_like(beta, np.nan)
theta0[valid] = np.arcsin(argument[valid]) - (np.pi/2 - beta[valid])

# プロット
plt.figure()

plt.plot(beta * 180/np.pi, theta0 * 180/np.pi, label=r'$\theta_0(\beta)$')

# カットオフを表示
plt.axvline(beta_cutoff * 180/np.pi, linestyle='--', label='cutoff')

# beta = theta_c（最大点）
plt.axvline(theta_c * 180/np.pi, linestyle=':', label='beta = theta_c')

plt.xlabel(r'$\beta$ (deg)')
plt.ylabel(r'$\theta_0$ (deg)')
plt.title('Emission angle vs taper angle')
plt.legend()
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 屈折率
n1 = 1.447
n2 = 1.431
n0 = 1.0

# 臨界角
theta_c1 = np.arcsin(n2 / n1)
theta_c2 = np.arcsin(n0 / n1)

# パラメータ範囲
theta1 = np.linspace(theta_c1 * 1.01, np.pi/2 - 1e-3, 200)
beta = np.linspace(0, np.pi/2, 200)

THETA1, BETA = np.meshgrid(theta1, beta)

# phi1
phi1 = THETA1 - BETA

# 条件1：コア–クラッド界面での全反射
cond1 = THETA1 > theta_c1

# 条件2：出射面での全反射を防止
cond2 = phi1 < theta_c2

# スネルの法則の定義域
argument = (n1 / n0) * np.sin(phi1)
cond3 = argument <= 1

# 全条件
valid = cond1 & cond2 & cond3

# theta0 を計算
theta0 = np.full_like(THETA1, np.nan)
theta0[valid] = np.arcsin(argument[valid]) - phi1[valid]

# 最大値を探索
max_idx = np.nanargmax(theta0)
max_theta0 = np.nanmax(theta0)

beta_max = BETA.flatten()[max_idx]
theta1_max = THETA1.flatten()[max_idx]

# プロット
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(
    BETA * 180/np.pi,
    THETA1 * 180/np.pi,
    theta0 * 180/np.pi,
    cmap='viridis'
)

# 最大点を表示
ax.scatter(
    beta_max * 180/np.pi,
    theta1_max * 180/np.pi,
    max_theta0 * 180/np.pi,
    color='red',
    s=50,
    label='Max'
)

ax.set_xlabel(r'$\beta$ (deg)')
ax.set_ylabel(r'$\theta_1$ (deg)')
ax.set_zlabel(r'$\theta_0$ (deg)')
ax.set_title('3D map of emission angle')

ax.legend()
plt.show()

# 最大値を出力
print("Max theta0 (deg):", max_theta0 * 180/np.pi)
print("At beta (deg):", beta_max * 180/np.pi)
print("At theta1 (deg):", theta1_max * 180/np.pi)
```
