# TE mode and Bessel functions
![Fig. 6](../.figure/fig6.jpg)
## 1. 좌표계와 기호

원통좌표: $
(r,\phi,z)
$

- $z$: 광축
- $r$: 광축으로부터 거리
- $\phi$: 횡단면에서의 방위각 (azimuth)


광선의 진행방향:

$$
\mathbf{k} = k_r \hat r + \beta \hat z
$$

여기서$
\theta = \angle(\mathbf{k},\hat z)
$를 정의하자.

즉

$$
k_r=nk_0\sin\theta
$$

$$
\beta=nk_0\cos\theta
$$

이다. 즉 $\theta$는 광선이 광축에 대해 얼마나 기울어졌는지를 나타낸다.

---

## 2. TE 모드란?

TE는

$$
E_z=0
$$

이다. 즉 전기장이 광축 방향 성분을 갖지 않는다. 하지만 $\frac{\partial}{\partial z}\neq0$일 수 있으므로 빛은 여전히 z방향으로 진행한다.

예를 들어 $x$방향으로 전기장이 진동하는 자유공간 평면파 $\mathbf E = E_0\hat x e^{i(kz-\omega t)}$ 는 $E_z=0$이지만 z방향으로 전파된다.

## 3. 광선 모델

도파관 내부의 한 평면파 성분:

$$
\mathbf{k}
=
k_r\hat r
+
\beta\hat z
$$


균일매질의 단일 평면파 조건: $\mathbf E\cdot \mathbf k=0$ 이므로 $\mathbf E \parallel \hat\phi$를 선택할 수 있다. 왜냐하면

$$
\hat\phi\perp \hat r
$$

$$
\hat\phi\perp \hat z
$$

이기 때문이다. 따라서 $ \mathbf{k} = k_r\hat r + \beta\hat z$ 를 따라 전파되는 평면파 하나만 보면 전기장의 진동방향 $\mathbf E$ 의 방향은 $\mathbf \phi$ 방향과 동일하다.

---

## 4. 모드 형성

광선이 벽에서 반사되면 $k_r \rightarrow -k_r$가 된다. 하지만 $ \beta $는 유지된다. 즉 $
\theta
$의 크기는 유지되고 부호만 바뀐다.

그래서$
+k_r
$광선과$
-k_r
$광선이 동시에 존재할때 이 두 파를 합치면

$$
e^{i(k_r r+\beta z)}
+
e^{i(-k_r r+\beta z)}
$$

$$
=
2\cos(k_r r)e^{i\beta z}
$$

이다.


중요한 점: $
\langle k_r\rangle=0
$이다. 즉 평균 횡방향 운동량은 없다.하지만 $k_r$ 자체는 남는다.그래서 횡방향 모드 구조가 형성된다. 즉 
$$
\boxed{\langle k_r\rangle=0}
$$ 
$$
\boxed{k_r\neq0}
$$.


전파각도 $\theta$ 를 따라 전파되는 평면파를 모든 방위각 $\psi$ 에 대해서 적분하면 원형 대칭 모드에서

$$
E(r,\phi,\theta) = \int_0^{2\pi}
e^{i k_r r\cos(\phi-\psi)}
\,d\psi
$$

와 같은 적분에서 나온다.

여기서

- $\phi$: 관측점의 방위각
- $r$: 관측점의 반경

이다.



베셀 함수의 적분 표현:

$$
J_0(k_r r)
=
\frac{1}{2\pi}
\int_0^{2\pi}
e^{i k_r r\cos(\phi-\psi)}
\,d\psi
$$

따라서

$$
\boxed{
\int_0^{2\pi}
e^{i k_r r\cos(\phi-\psi)}
\,d\psi
=
2\pi J_0(k_r r)
}
$$

이다. 그런데 $J_l$ 는 언제 나오는가?

각 평면파에 위상 가중치 $
e^{i l\psi}
$ 를 주면

$$
\boxed{
\int_0^{2\pi}
e^{i l\psi}
e^{i k_r r\cos(\phi-\psi)}
\,d\psi = 
2\pi i^l
J_l(k_r r)
e^{i l\phi}
}
$$

가 된다.



$TE_{0m}$에서는 $
H_z
\propto
J_0(k_r r)
$ 이다. 맥스웰 방정식으로부터 $
E_\phi
\propto
-\frac{\partial H_z}{\partial r}
$ 이므로

$$
E_\phi
\propto
-\frac{\partial}{\partial r}
J_0(k_r r)
$$

이다.

베셀 함수 미분 공식

$$
\frac{dJ_0(x)}{dx}
=
-J_1(x)
$$

를 사용하면

$$
E_\phi
\propto
J_1(k_r r)
$$

를 얻는다.

## 5. $l$의 의미

여기서 가장 중요한 부분. 모드는 일반적으로 $
e^{il\phi}
$ 를 포함한다. 즉
$$
E(r,\phi,z)=J_l(k_r r)e^{i l \phi}e^{-i\beta z}
$$

여기서 $
\phi
$ 는 횡단면 방위각이다.



$
l=0
$ 이면 $
e^{i0\phi}=1
$ 이다. 즉 위상이 \(\phi\)에 의존하지 않는다.

$
l=1
$ 이면 $
e^{i\phi}
$이다. 한 바퀴 돌면 위상이 $2\pi$ 변한다.


$
l=2
$ 이면 $
e^{i2\phi}
$ 이다. 한 바퀴 돌면 위상이 $4\pi$변한다.


따라서 $l$ 은 **편광 방향이 아니라 위상 감김수(topological charge)** 를 나타낸다.
