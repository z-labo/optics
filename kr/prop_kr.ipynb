# Wave propagation: Fresnel and Fraunhofer diffraction

![Fig. 8](../.figure/fig8.jpg)

## 1. 좌표계와 기본 개념

파동의 자유공간 전파를 생각한다. 개구면(aperture plane)을 $z=0$ 평면으로 두고, 그 위의 좌표를

$$
(\xi,\eta)
$$

라고 하자. 관측면(observation plane)은 개구면으로부터 $z$만큼 떨어진 평면이며, 그 위의 좌표를

$$
(x,y)
$$

라고 둔다.

개구면 직후의 복소수 진폭을

$$
u_0(\xi,\eta)
$$

라고 하고, 거리 $z$만큼 전파한 뒤 점 $P$ 에서의 복소수 진폭을

$$
u_P(x,y;z)
$$

라고 쓴다.

스칼라로 표시한 파수는 $k=\frac{2\pi}{\lambda}$ 이다. 여기서 $\lambda$는 파장이다.

개구면 위의 한 점 $(\xi,\eta,0)$에서 관측점 $P(x,y,z)$까지의 거리는

$$
r=\sqrt{z^2+(x-\xi)^2+(y-\eta)^2}
$$

이다.

Huygens의 원리에 따르면, 개구면 위의 각 점은 2차 구면파를 방출하는 점광원처럼 작용한다. 따라서 관측면의 장은 개구면 위 모든 점에서 나온 구면파들의 중첩으로 주어진다.

다만 다음과 같이 구분해서 이해하는 것이 정확하다.

- 가까운 거리에서는 각 점광원에서 나온 구면파들의 위상 차이를 그대로 고려해야 한다. 이 영역이 Fresnel diffraction 영역이다.
- 충분히 먼 거리에서는 점광원이 이루는 파면을 평면파로 보고, 각 점광원들이 형성하는 평면파들이 관측 방향별로 위상 변화가 거의 선형적으로 정리된다. 이 영역이 Fraunhofer diffraction 영역이다.

## 2. Huygens-Kirchhoff integral

개구면 이전의 파동 형성 과정은 고려하지 않고, 개구면 직후의 복소수 진폭 $u_0(\xi,\eta)$가 주어졌다고 하자.

Huygens-Kirchhoff diffraction integral의 한 형태는 다음과 같이 쓸 수 있다.

$$
u(x,y;z)
=
-\frac{ik}{4\pi}
\iint
u_0(\xi,\eta)
\frac{e^{ikr}}{r}
\left(1+\cos\theta\right)
\,d\xi d\eta .
$$

여기서 $\theta$는 개구면의 법선 방향과 관측 방향이 이루는 각도이다. 또한

$$
k=\frac{2\pi}{\lambda}
$$

이므로 위 식은

$$
u(x,y;z)
=
-\frac{i}{2\lambda}
\iint
u_0(\xi,\eta)
\frac{e^{ikr}}{r}
\left(1+\cos\theta\right)
\,d\xi d\eta
$$

로 쓸 수 있다.

경사인자(obliquity factor)를

$$
K(\theta)=\frac{1+\cos\theta}{2}
$$

라고 두면,

$$
u(x,y;z)
=
-\frac{i}{\lambda}
\iint
u_0(\xi,\eta)
\frac{e^{ikr}}{r}
K(\theta)
\,d\xi d\eta .
$$

paraxial approximation, 즉 관측각 $\theta$가 충분히 작은 경우에는

$$
\cos\theta \approx 1
$$

이므로

$$
K(\theta)\approx 1
$$

이다. 따라서

$$
u(x,y;z)
\approx
-\frac{i}{\lambda}
\iint
u_0(\xi,\eta)
\frac{e^{ikr}}{r}
\,d\xi d\eta .
$$

또한 진폭 항에 들어 있는 $r$은 위상항 $e^{ikr}$에 비해 변화가 느리므로

$$
r \approx z
$$

로 둘 수 있다. 그러면

$$
u(x,y;z)
\approx
-\frac{i}{\lambda z}
\iint
u_0(\xi,\eta)
e^{ikr}
\,d\xi d\eta .
$$

여기서 중요한 점은 다음과 같다.

진폭에 해당하는 분모의 $r$은 $z$로 근사할 수 있지만, 지수함수 안에 있는 $r$은 함부로 $z$로 바꾸면 안 된다. 왜냐하면 광학에서는 $k=2\pi/\lambda$가 매우 크기 때문에, 아주 작은 거리 차이도 큰 위상 차이를 만들 수 있기 때문이다.

## 3. 거리 $r$의 Taylor expansion

이제 지수함수 안에 있는

$$
r=\sqrt{z^2+(x-\xi)^2+(y-\eta)^2}
$$

를 근사해 보자.

먼저

$$
r
=
z
\sqrt{
1+
\frac{(x-\xi)^2+(y-\eta)^2}{z^2}
}
$$

로 쓸 수 있다.

여기서

$$
\sqrt{1+s}
=
1+\frac{s}{2}-\frac{s^2}{8}+\cdots
$$

를 사용하면,

$$
r
=
z
\left[
1
+
\frac{1}{2}
\frac{(x-\xi)^2+(y-\eta)^2}{z^2}
-
\frac{1}{8}
\left(
\frac{(x-\xi)^2+(y-\eta)^2}{z^2}
\right)^2
+
\cdots
\right].
$$

따라서

$$
r
\approx
z
+
\frac{(x-\xi)^2+(y-\eta)^2}{2z}
$$

가 된다.

주의할 점은, 단순히 Taylor expansion의 뒤쪽 항이 작아 보인다고 해서 바로 버릴 수 있는 것은 아니라는 점이다. 실제로 버려지는 항이 만드는 위상 오차

$$
k\Delta r
$$

가 충분히 작아야 Fresnel approximation이 성립한다.

이 근사를 Huygens-Kirchhoff integral에 대입하면,

$$
e^{ikr}
\approx
e^{ikz}
\exp
\left[
i\frac{k}{2z}
\left\{
(x-\xi)^2+(y-\eta)^2
\right\}
\right].
$$

따라서

$$
u(x,y;z)
\approx
-\frac{i e^{ikz}}{\lambda z}
\iint
u_0(\xi,\eta)
\exp
\left[
i\frac{k}{2z}
\left\{
(x-\xi)^2+(y-\eta)^2
\right\}
\right]
\,d\xi d\eta .
$$

$k=2\pi/\lambda$이므로,

$$
\frac{k}{2z}
=
\frac{\pi}{\lambda z}
$$

이다. 따라서 Fresnel diffraction integral은

$$
u(x,y;z)
\approx
-\frac{i e^{ikz}}{\lambda z}
\iint
u_0(\xi,\eta)
\exp
\left[
i\frac{\pi}{\lambda z}
\left\{
(x-\xi)^2+(y-\eta)^2
\right\}
\right]
\,d\xi d\eta .
$$

또는

$$
\frac{1}{i}=-i
$$

를 이용하여

$$
u(x,y;z)
\approx
\frac{e^{ikz}}{i\lambda z}
\iint
u_0(\xi,\eta)
\exp
\left[
i\frac{\pi}{\lambda z}
\left\{
(x-\xi)^2+(y-\eta)^2
\right\}
\right]
\,d\xi d\eta
$$

라고 써도 같은 식이다.

## 4. Fresnel diffraction integral의 Fourier transform 형태

이제 지수항 안의 제곱을 전개한다.

$$
(x-\xi)^2+(y-\eta)^2
=
x^2+y^2+\xi^2+\eta^2-2x\xi-2y\eta .
$$

따라서

$$
u(x,y;z)
\approx
\frac{e^{ikz}}{i\lambda z}
\iint
u_0(\xi,\eta)
\exp
\left[
i\frac{\pi}{\lambda z}
\left(
x^2+y^2+\xi^2+\eta^2-2x\xi-2y\eta
\right)
\right]
\,d\xi d\eta .
$$

여기서 $x$와 $y$만 포함하는 항은 적분 변수 $\xi,\eta$와 무관하므로 적분 밖으로 꺼낼 수 있다.

$$
u(x,y;z)
\approx
\frac{e^{ikz}}{i\lambda z}
\exp
\left[
i\frac{\pi}{\lambda z}
(x^2+y^2)
\right]
\iint
u_0(\xi,\eta)
\exp
\left[
i\frac{\pi}{\lambda z}
(\xi^2+\eta^2)
\right]
\exp
\left[
-i\frac{2\pi}{\lambda z}
(x\xi+y\eta)
\right]
\,d\xi d\eta .
$$

이 식이 Fresnel diffraction integral의 중요한 형태이다.

임의의 함수 $g(\xi,\eta)$의 Fourier transform을 다음과 같이 정의하자.

$$
\mathcal{F}\{g(\xi,\eta)\}(f_x,f_y)
=
\iint
g(\xi,\eta)
e^{-i2\pi(f_x\xi+f_y\eta)}
\,d\xi d\eta .
$$

그러면 Fresnel diffraction은

$$
g(\xi,\eta)
=
u_0(\xi,\eta)
\exp
\left[
i\frac{\pi}{\lambda z}
(\xi^2+\eta^2)
\right]
$$

의 Fourier transform으로 볼 수 있다. 이때 Fourier transform의 spatial frequency는

$$
f_x=\frac{x}{\lambda z},
\qquad
f_y=\frac{y}{\lambda z}
$$

이다.

따라서 Fresnel diffraction은 단순한 Fourier transform이 아니라,

$$
u_0(\xi,\eta)
$$

에 2차 위상항(chirp phase)

$$
\exp
\left[
i\frac{\pi}{\lambda z}
(\xi^2+\eta^2)
\right]
$$

을 곱한 뒤 Fourier transform한 형태이다.

## 5. Fraunhofer diffraction

Fraunhofer diffraction은 Fresnel diffraction에서 한 단계 더 근사한 것이다.

Fresnel integral 안에는 다음과 같은 개구면 쪽 2차 위상항이 있다.

$$
\exp
\left[
i\frac{\pi}{\lambda z}
(\xi^2+\eta^2)
\right].
$$

Fraunhofer approximation에서는 이 항의 위상 변화가 개구면 전체에서 충분히 작다고 가정한다.

즉,

$$
\frac{\pi}{\lambda z}(\xi^2+\eta^2) \ll 1
$$

이면

$$
\exp
\left[
i\frac{\pi}{\lambda z}
(\xi^2+\eta^2)
\right]
\approx 1
$$

로 둘 수 있다.

개구의 대표적인 크기를 $D$라고 하면, 대략적으로

$$
z \gg \frac{D^2}{\lambda}
$$

일 때 Fraunhofer approximation이 잘 성립한다.

이 근사를 적용하면,

$$
u_F(x,y;z)
\approx
\frac{e^{ikz}}{i\lambda z}
\exp
\left[
i\frac{\pi}{\lambda z}
(x^2+y^2)
\right]
\iint
u_0(\xi,\eta)
\exp
\left[
-i\frac{2\pi}{\lambda z}
(x\xi+y\eta)
\right]
\,d\xi d\eta .
$$

Fourier transform의 정의와 비교하면,

$$
u_F(x,y;z)
\approx
\frac{e^{ikz}}{i\lambda z}
\exp
\left[
i\frac{\pi}{\lambda z}
(x^2+y^2)
\right]
\mathcal{F}\{u_0(\xi,\eta)\}
\left(
\frac{x}{\lambda z},
\frac{y}{\lambda z}
\right).
$$

즉, Fraunhofer diffraction에서 관측면의 복소수 진폭은 개구면 복소수 진폭의 Fourier transform에 비례한다.

관측되는 세기는

$$
I_F(x,y;z)
=
|u_F(x,y;z)|^2
$$

이므로, 앞에 곱해진 순수 위상항은 세기에 영향을 주지 않는다. 따라서 Fraunhofer 영역의 intensity pattern은

$$
I_F(x,y;z)
\propto
\left|
\mathcal{F}\{u_0(\xi,\eta)\}
\left(
\frac{x}{\lambda z},
\frac{y}{\lambda z}
\right)
\right|^2
$$

로 주어진다.


단, 이 관계에서 Fourier transform의 좌표는 물리적 관측 좌표 $(x,y)$가 아니라

$$
\left(
\frac{x}{\lambda z},
\frac{y}{\lambda z}
\right)
$$

로 스케일된 spatial frequency 좌표이다.
